```python
from agno.agent import Agent
from agno.db.sqlite import SqliteDb
from agno.models.openai import OpenAIChat
from agno.models.xai import xAI
from agno.tools.file import FileTools
from dotenv import load_dotenv
import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../..")))

load_dotenv()

from Backend.Tools.Allocation_tools import (
    present_value_calculator,
    portfolio_statistics,
    simple_optimizer,
    stress_tester,
)
def get_data():
    with open("D:/CIMR-OS/Backend/Inputs/Allocation_Input.json", "r") as f:
        return f.read()
ActuarialOptimizer = Agent(
    name="Actuarial Optimizer",
    model=xAI(id="grok-3-mini", api_key=os.getenv("XAI_API_KEY")),
    description="Optimizes CIMR portfolio allocation using actuarial projections, market data, and risk constraints.",
    instructions="""

   use the get_data function to retrieve the data you have to work with, if the user doesn't provide a custom data.

    You can perform actuarial tasks using your tools:
    - To compute present value of liabilities, use present_value_calculator.
    - To compute portfolio statistics, use portfolio_statistics.
    - To propose a new allocation, use simple_optimizer.
    - To run stress tests, use stress_tester.
    Always expect JSON as input, and return JSON as output.
    """,
    tools=[present_value_calculator, portfolio_statistics, simple_optimizer, stress_tester, FileTools(), get_data],
)
```
